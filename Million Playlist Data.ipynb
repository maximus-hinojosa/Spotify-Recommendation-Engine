```python
import pandas as pd
import numpy as py
import Recommenders as Recommenders
```

```python
#Listen count for unique track per user
#Song_id is common key
song_df_1 = pd.read_csv('triplets_file.csv')
song_df_1.head()
```

```python
#Data about each unique track
#Song_id is primary key
song_df_2 = pd.read_csv('song_data.csv')
song_df_2.head()
```

```python
#Combine both datasets into single data frame
song_df = pd.merge(song_df_1, song_df_2.drop_duplicates(['song_id']), on='song_id', how='left')
song_df.head()
```

```python
len(song_df_1) #2 million entries of song listens
```

```python
len(song_df_2) #1 million unique songs
```

# Data Pre-processing

```python
song_df['song'] = song_df['title']+ "-" + song_df['artist_name'] #New column combining artist and song
song_df.head()
```

```python
#Aggregate sum of listen count of the songs
song_grouped = song_df.groupby(['song']).agg({'listen_count':'count'}).reset_index()
song_grouped.head()
song_grouped.sort_values(['listen_count', 'song'], ascending=[0,10000000]) 
```

```python
grouped_sum = song_grouped['listen_count'].sum() #Sum of listen Count
song_grouped['percentage'] = ( song_grouped['listen_count'] / grouped_sum )  #Percentage of total listening share by song
song_grouped.sort_values(['listen_count', 'song'], ascending=[0,1]) 
```
